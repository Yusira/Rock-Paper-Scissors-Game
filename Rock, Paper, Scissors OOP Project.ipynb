```python
#defining a class 'Game' that will initiate with the player's name
class Game:
    def __init__(self, name):
        self.name = name
        #setting both the player's and computer's score to 0
        self.pscore = 0
        self.cscore = 0
        self.welcome_page()
        
    def welcome_page(self):
        print(f"Welcome {self.name} to the Rock, Paper, Scissors Game\n\nGame Menu:\n\n1.) Game Instructions \n2.) Play Game\n")
        menu = int(input("Choose an option: "))
        
        #coding the menu option for the player
        if menu == 1:
            self.game_instructions()
        elif menu == 2:
            self.play_game()
        else:
            print("Invalid Option")
        
    def game_instructions(self):
        print("\nInstructions: \n\nThe rules are simple.\n\nChoose a weapon, 'R' for Rock, 'S' for Scissors, or 'P' for Paper.\nYour opponent(the computer) will do the same.\n\n - Rock will crush Scissors\n - Scissors will cut Paper\n - Paper will beat Rock, the mechanism behind that is not completely understood\n\nNow play the game and enjoyyyyy...")
        
    def play_game(self):
        import random
        print("\n\nWelcome to the Single Player Mode")
        
        #possible options for the computer to randomly select
        options = ["Rock", "Paper", "Scissors"]
        
        #constructing a loop for the game to run 5 times
        for i in range(1,6):
            #computer's choice
            comp_play = random.choice(options)
            
            print(f"\nRound {i}!\n")
            
            print("Weapons: \n'R'ock \n'P'aper \n'S'cissors")
            
            #player's choice
            play = input("Choose your Weapon (R, P, S):")
            
            #conditional statements for all the possible game-plays between the player and the computer
            if play == 'R' and comp_play == 'Rock':
                print(f"\n\nYour opponent chose: {comp_play}")
                print("\nIt was a tie\n")
                print(f"Your Score - {self.get_pscore()}\nOpponent's Score - {self.get_cscore()}")
            elif play == 'R' and comp_play == 'Scissors':
                print(f"\n\nYour opponent chose: {comp_play}")
                print("\nYou Won!!\n")
                self.set_pscore()
                print(f"Your Score - {self.get_pscore()}\nOpponent's Score - {self.get_cscore()}")
            elif play == 'R' and comp_play == 'Paper':
                print(f"\n\nYour opponent chose: {comp_play}")
                print("\nYou Lost!!\n")
                self.set_cscore()
                print(f"Your Score - {self.get_pscore()}\nOpponent's Score - {self.get_cscore()}")
            elif play == 'S' and comp_play == "Rock":
                print(f"\n\nYour opponent chose: {comp_play}")
                print("\nYou Lost!!\n")
                self.set_cscore()
                print(f"Your Score - {self.get_pscore()}\nOpponent's Score - {self.get_cscore()}")
            elif play == 'S' and comp_play == "Scissors":
                print(f"\n\nYour opponent chose: {comp_play}")
                print("\nIt was a tie\n")
                print(f"Your Score - {self.get_pscore()}\nOpponent's Score - {self.get_cscore()}")
            elif play == 'S' and comp_play == "Paper":
                print(f"\n\nYour opponent chose: {comp_play}")
                print("\nYou Won!!\n")
                self.set_pscore()
                print(f"Your Score - {self.get_pscore()}\nOpponent's Score - {self.get_cscore()}")
            elif play == 'P' and comp_play == "Rock":
                print(f"\n\nYour opponent chose: {comp_play}")
                print("\nYou Won!!\n")
                self.set_pscore()
                print(f"Your Score - {self.get_pscore()}\nOpponent's Score - {self.get_cscore()}")
            elif play == 'P' and comp_play == "Scissors":
                print(f"\n\nYour opponent chose: {comp_play}")
                print("\nYou Lost!!\n")
                self.set_cscore()
                print(f"Your Score - {self.get_pscore()}\nOpponent's Score - {self.get_cscore()}")
            elif play == 'P' and comp_play == "Paper":
                print(f"\n\nYour opponent chose: {comp_play}")
                print("\nIt was a tie\n")
                print(f"Your Score - {self.get_pscore()}\nOpponent's Score - {self.get_cscore()}")
            else:
                print("Choose Again!!")
        
        #final statements at the end of the 5th round
        if self.pscore > self.cscore:
            print("\nCONGRATULATIONS, YOU WON THE GAME\n\nYippeeeeee!!!!!")
        elif self.pscore < self.cscore:
            print("\nSORRY, YOU LOST THE GAME\n\nTRY AGAIN!!!")
        else:
            print("\nIT WAS A TIE!!!")
            
    
    #defining a function that increases the player's score by 10 for each win
    def set_pscore(self):
        self.pscore = self.pscore + 10
    
    #defining a function that increases the computer's score by 10 for each win
    def set_cscore(self):
        self.cscore = self.cscore + 10
        
    #defining a function that returns the player's score
    def get_pscore(self):
        return self.pscore
    
    #defining a function that returns the computer's score
    def get_cscore(self):
        return self.cscore
```
